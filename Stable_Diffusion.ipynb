```python
import torch
from diffusers import StableDiffusionPipeline
model_id='runwayml/stable-diffusion-v1-5' #model
pipe=StableDiffusionPipeline.from_pretrained(model_id)
pipe=pipe.to('cuda') #cuda-->graphics
```

```python
prompt='a jupyter planet'
image=pipe(
    prompt,
    height=512,
    weight=512,
).images[0] #many images are generated,out of which we only require one(the first one)
image
```

```python
#function
def teximg(prompt):
  image=pipe(prompt,guidance_scale=9,height=769,weight=769).images[0] #many images are generated,out of which we only require one(the first one)
  return image
```

```python
img=teximg('football in swimming pool')
img
```

```python
#User Interface
import gradio as gr
demo=gr.Interface(fn=teximg,inputs='text',outputs='image')
demo.launch()
```
